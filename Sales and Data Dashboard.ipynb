## Sales data Dashboard
This notebook analyzes sales data for a fictional e-commerce store and builds a simple dashboard with insights.

## ðŸ§© Problem Statement

Analyze product sales to identify:
- Monthly revenue trends
- Top-selling products
- Most profitable regions

## ðŸ“¥ Dataset

We'll use a CSV file with the following columns:
- `Date`
- `Region`
- `Product`
- `Quantity Sold`
- `Unit Price`

## ðŸ§¹ Step 1: Load and Clean Data

- Read the CSV using `pandas`
- Convert `Date` column to datetime
- Handle missing values if needed

## ðŸ“Š Step 2: Analysis

- Total revenue per month
- Top 5 selling products
- Best performing region
- Bar chart: product vs. revenue
- Line plot: monthly trend

## ðŸ“ˆ Step 3: Visualizations

Use `matplotlib` or `seaborn` to create:
- Line graph (Revenue over months)
- Bar chart (Top products)
- Pie chart (Revenue by region)


## ðŸ§  Conclusion

Summarize 3â€“4 key business insights from the data.

```python
# ðŸ“Š Sales Data Dashboard

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('sales_data.csv')
df['Date'] = pd.to_datetime(df['Date'])

# Show the first few rows
df.head()
# Check data types and nulls
df.info()

# Add Month column
df['Month'] = df['Date'].dt.to_period('M')

# Total Revenue per month
monthly_revenue = df.groupby('Month')['Revenue'].sum()
monthly_revenue.plot(kind='line', figsize=(10,5), title='Monthly Revenue')
plt.ylabel("Revenue")
plt.show()
top_products = df.groupby('Product')['Revenue'].sum().sort_values(ascending=False)
top_products.plot(kind='bar', title='Top Selling Products')
plt.ylabel("Revenue")
plt.show()
region_revenue = df.groupby('Region')['Revenue'].sum()
region_revenue.plot(kind='pie', autopct='%1.1f%%', title='Revenue by Region')
plt.ylabel("")
plt.show()
```

## Insights:
- Highest revenue was in [Month]
- Best-selling product was [Product]
- [Region] contributed the most to total sales
